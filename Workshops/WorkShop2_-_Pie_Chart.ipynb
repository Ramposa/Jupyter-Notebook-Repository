```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

t_mov=pd.read_csv('types_movies.csv') # Defined as t_mov
```

```python
plt.pie(t_mov['Percentage'],
       labels=t_mov['Sector'])

plt.axis('equal')

plt.show()
```

```python
wedge, text, autotexts = plt.pie(t_mov['Percentage'],
                                labels=t_mov['Sector'],
                                autopct='%.2f')

plt.axis('equal')

print('Wedge:', wedge) # Wedges: A list of patch object representing each wedge
print('Text: ', text) # Texts: List of text object represetnign the labels.
print('Autotexts: ', autotexts) # Autotexts:  List of text object for numeric values - this is vaiable for autopct value for pie
# chart is NONE.
```

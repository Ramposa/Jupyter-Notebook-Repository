```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

stock_data = pd.read_csv('stocks.csv')
```

```python
stock_data.head()
```

```python
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.head()
#Convert Date to pandas DateTime for easier plotting of data.
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.set_title('AAPL vs IBM(insert)')
ax1.plot(stock_data['Date'],
        stock_data['AAPL'],
        color='green')

ax2.plot(stock_data['Date'],
        stock_data['IBM'],
        color='blue')

plt.show
#Comapre Stocks side-by-side
```

```python
# Figsize for width and height of plot
# Stocks comparisons

fig= plt.figure(figsize=(15,7))
fig.suptitle('Stock Price Comparision 2007-2017',
            fontsize=20)

ax1 = fig.add_subplot(231)
ax1.set_title('MSFT')
ax1.plot(stock_data['Date'],
        stock_data['MSFT'],
        color='green')

ax2 = fig.add_subplot(232)
ax2.set_title('NFLX')
ax2.plot(stock_data['Date'],
        stock_data['NFLX'],
        color='green')

ax3 = fig.add_subplot(233)
ax3.set_title('ORCL')
ax3.plot(stock_data['Date'],
        stock_data['ORCL'],
        color='green')

ax4 = fig.add_subplot(234)
ax4.set_title('GOOG')
ax4.plot(stock_data['Date'],
        stock_data['GOOG'],
        color='green')

ax5 = fig.add_subplot(235)
ax5.set_title('CVX')
ax5.plot(stock_data['Date'],
        stock_data['CVX'],
        color='green')

ax6 = fig.add_subplot(236)
ax6.set_title('ADBE')
ax6.plot(stock_data['Date'],
        stock_data['ADBE'],
        color='green')

plt.show
```
